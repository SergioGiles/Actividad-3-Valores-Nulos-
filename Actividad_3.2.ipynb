```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  

df = pd.read_csv('PortoPortugal_50.csv')
```

```python
numeric_columns = df.select_dtypes(include=['int', 'float']).columns
```

```python
# a) Reemplazar valores nulos con el método de “mediana” en las columnas tipo “numeric (int y float)”


def reemplazar_nulos_con_mediana(df):
    
    columnas_numericas = df.select_dtypes(include=['int', 'float']).columns
    
    for columnas in columnas_numericas:
        df[columnas].fillna(df[columnas].median(), inplace=True)
    
    df.to_csv('reemplazo_mediana.csv', index=False)
    return df

df_mediana = reemplazar_nulos_con_mediana(df)
```

```python
# b) Reemplazar valores nulos con el método de “ffill” en las columnas tipo “object, datetime y category”

def reemplazar_nulos_con_ffill(df):
    
    columnas_no_numericas = df.select_dtypes(include=['object', 'datetime', 'category']).columns
    
    for col in columnas_no_numericas:
        df[col].fillna(method='ffill', inplace=True)
    
    df.to_csv('reemplazo_ffill.csv', index=False)
    return df

df_ffill = reemplazar_nulos_con_ffill(df)
```

```python
# c) Reemplazar valores nulos con el método de “bfill” en las columnas tipo “object, datetime y category”

def reemplazar_nulos_con_bfill(df):

    columnas_no_numericas = df.select_dtypes(include=['object', 'datetime', 'category']).columns
    
    for col in columnas_no_numericas:
        df[col].fillna(method='bfill', inplace=True)
    
    df.to_csv('reemplazo_bfill.csv', index=False)
    return df

df_bfill = reemplazar_nulos_con_bfill(df)
```

```python
# d) Reemplazar valores nulos con el método de “string concreto” en las columnas tipo “object, datetime y category”


def reemplazar_nulos_con_string(df, string_concreto='NO'):
    
    columnas_no_numericas = df.select_dtypes(include=['object', 'datetime', 'category']).columns
    
    # Reemplazar los valores nulos con un string concreto
    for col in columnas_no_numericas:
        df[col].fillna(string_concreto, inplace=True)
    
    # Guardar el resultado en un nuevo archivo CSV
    df.to_csv('reemplazo_string_concreto.csv', index=False)
    return df

# Llamar a la función
df_string_concreto = reemplazar_nulos_con_string(df, string_concreto='NO')
```

```python
# e) Reemplazar valores nulos con el método de “constante” en las columnas tipo “numeric (int y float)”

def reemplazar_nulos_con_constante(df, constante=0):
    
    columnas_numericas = df.select_dtypes(include=['int', 'float']).columns
    
    for col in columnas_numericas:
        df[col].fillna(constante, inplace=True)
    
    df.to_csv('reemplazo_constante.csv', index=False)
    return df

df_constante = reemplazar_nulos_con_constante(df, constante=0)
```

```python
df_mediana = reemplazar_nulos_con_mediana(df)
df_ffill = reemplazar_nulos_con_ffill(df)
df_bfill = reemplazar_nulos_con_bfill(df)
df_string_concreto = reemplazar_nulos_con_string(df, string_concreto='Desconocido')
df_constante = reemplazar_nulos_con_constante(df, constante=0)
```
